**<font size=6 color=gold> Characteristic function (CF) of selected discrete (frequency) distributions </font>** 

```r
# devtools::install_github("gajdosandrej/CharFunToolR") 
library(CharFunToolR)
```

<h2>BINOMIAL DISTIBUTION</h2>

`cfN_Binomial}(t,n,p)` evaluates the characteristic function $\text{cf(t)}$ of the Binomial distribution with the parameters `n` (number of trials, $n\in\mathbb{N}$)
 and `p` (success probability, $p\in[0,1]$), i.e. 

 $$\text{cf(t)} = \text{cfN_Binomial(t,n,p)} = \left(1 - p + p\cdot e^{it}\right)^n$$

 For more details see WIMMER and ALTMANN (1999), also WIKIPEDIA:
 <https://en.wikipedia.org/wiki/Binomial_distribution>

 `cfN_Binomial(t,n,p,cfX)` evaluates the compound characteristic function:

 $$\text{cf(t)} = \text{cfN_Binomial(-1i*log(cfX(t)),n,p)},$$

 where $\text{cfX}$ is function handle of the characteristic function $\text{cfX(t)}$ of a
 continuous distribution and/or random variable X.  
 
 Note that such CF is characteristic function of the compound distribution,
 i.e. distribution of the random variable $ Y = X_1 + ... + X_N $, where $X_i \sim
 F_X$ are i.i.d. random variables with common CF $\text{cfX(t)}$, and $N \sim F_N $ is
 independent RV with its CF given by $\text{cfN(t)}$.  

$\textbf{a) EXAMPLE: CF of the Binomial distribution with n = 25, p = 0.3}$

```r
n <- 25
p <- 0.3
t <- seq(-15, 15, length.out = 1001)
plotReIm(function(t)
        cfN_Binomial(t, n, p),
        t,
        title = "CF of the Binomial distribution with n = 25, p = 0.3")
```

$\textbf{b) EXAMPLE: CF of the compound Binomial-Exponential distribution}$

```r
n <- 25
p <- 0.3
lambda <- 10
cfX <- function(t)
        cfX_Exponential(t, lambda)
t <- seq(-10, 10, length.out = 501)
plotReIm(function(t)
        cfN_Binomial(t, n, p, cfX),
        t,
        title = "CF of the compound Binomial-Exponential distribution")
```

$\textbf{c) EXAMPLE: PDF/CDF of the compound Binomial-Exponential distribution}$

```r
n <- 25
p <- 0.3
lambda <- 5
cfX <- function(t)
        cfX_Exponential(t, lambda)
cf <- function(t)
        cfN_Binomial(t, n, p, cfX)
x <- seq(0, 5, length.out = 101)
prob <- c(0.9, 0.95, 0.99)
options <- list()
options$isCompound <- TRUE
result <- cf2DistGP(cf, x, prob, options)
```

<h2>DELAPORTE DISTRIBUTION</h2>

`cfN_Delaporte(t,a,b,c)` evaluates the characteristic function $\text{cf(t)}$ of the
 Delaporte distribution with the parameters `a` (parameter of variable mean,
 $a > 0$), `b` (parameter of variable mean, $b > 0$ ), and `c` (fixed mean, $c >
 0$), i.e.

 $$\text{cf(t)} = \text{cfN_Binomial(t,n,p)} = \left(1 - p + pe^{it}\right)^n = \left(\dfrac{b}{1+b}\right)^a \cdot 
 \left(1-\dfrac{e^{it}}{b+1}\right)^{(-a)} \cdot e^{-c(1-e^{it})}; $$

 For more details see WIMMER and ALTMANN (1999), also WIKIPEDIA:
 <https://en.wikipedia.org/wiki/Delaporte_distribution>

$\textbf{a) EXAMPLE: CF of the Delaporte distribution with a=2.2, b=3.3, c=4}$

```r
a <- 2.2
b <- 3.3
c <- 4
t <- seq(-15, 15, length.out = 1001)
plotReIm(function(t)
        cfN_Delaporte(t, a, b, c),
        t,
        title = "CF of the Delaporte distribution with a=2.2, b=3.3, c=4")
```

$\textbf{b) EXAMPLE: CF of the compound Delaport-Exponential distribution}$

```r
a <- 2.2
b <- 3.3
c <- 4
lambda <- 10
cfX <- function(t)
        cfX_Exponential(t, lambda)
t <- seq(-10, 10, length.out = 501)
plotReIm(function(t)
        cfN_Delaporte(t, a, b, c, cfX),
        t,
        title = "CF of the compound Delaport-Exponential distribution")
```

$\textbf{c) EXAMPLE: PDF/CDF of the compound Delaport-Exponential distribution}$

```r
a <- 2.2
b <- 3.3
c <- 4
lambda <- 5
cfX <- function(t)
        cfX_Exponential(t, lambda)
cf <- function(t)
        cfN_Delaporte(t, a, b, c, cfX)
x <- seq(0, 4, length.out = 101)
prob <- c(0.9, 0.95, 0.99)
options <- list()
options$isCompound <- TRUE
result <- cf2DistGP(cf, x, prob, options)
```

<h2>GENERALIZED POISSON DISTRIBUTION</h2>

`cfN_GeneralizedPoisson(t,a,p)` evaluates the characteristic function cf(t) of the Generalized-Poisson distribution, with the parameters `a` (parameter
 of variable mean, $a > 0$), and `p` (success probability, $p$ in $[0, 1]$), i.e.

 $$\text{cf(t)} = \text{cfN_GeneralizedPoisson(t,a,p)} = exp\left(a\left(\displaystyle\sum_{j=1}^\infty \left((pj)^{j-1}\frac{e^{-pj}}{j!}\right)e^{itj}-1\right)\right);$$ 

 The Generalized-Poisson distribution is equivalent with the Borel-Tanner
 distribution with parameters $(p, m)$, see WIMMER and ALTMANN (1999), where 
 $m \sim Poisson(a)$.

 `cfN_GeneralizedPoisson(t,a,p,cfX)` evaluates the compound characteristic function 

 $$\text{cf(t)} = \text{cfN_GeneralizedPoisson(-1i*log(cfX(t)),a,p)},$$

 where `cfX` is function handle of the characteristic function $\text{cfX(t)}$ of a
 continuous distribution and/or random variable $X$.  


$\textbf{a) EXAMPLE: CF of the GeneralizedPoisson distribution, a=10, p=0.5}$

```r
a <- 10
p <- 0.5
t <- seq(-10, 10, length.out = 501)
plotReIm(function(t)
        cfN_GeneralizedPoisson(t, a, p), t,
        title = "CF of the Generalized-Poisson distribution with a = 10, p = 0.5")
```

$\textbf{b) EXAMPLE: CF compound GeneralizedPoisson-Exponential distribution}$

```r
a <- 10
p <- 0.5
lambda <- 5
cfX <- function(t)
        cfX_Exponential(t, lambda)
t <- seq(-10, 10, length.out = 501)
plotReIm(function(t)
        cfN_GeneralizedPoisson(t, a, p, cfX), t,
        title = "CF of the compound Generalized-Poisson-Exponential distribution")
```

$\textbf{c) EXAMPLE: PDF/CDF of the compound GeneralizedPoisson-Exponential distribution}$

```r
a <- 10
p <- 0.5
lambda <- 5
cfX <- function(t)
        cfX_Exponential(t, lambda)
cf <- function(t)
        cfN_GeneralizedPoisson(t, a, p, cfX)
x <- seq(0, 15, length.out = 101)
prob <- c(0.9, 0.95, 0.99)
options <- list()
options$isCompound <- TRUE
result <- cf2DistGP(cf, x, prob, options)
```

 <h2>NEGATIVE BINOMIAL DISTRIBUTION</h2>

 `cfN_NegativeBinomial}(t,r,p)` evaluates the characteristic function $\text{cf(t)}$
 of the Negative-Binomial distribution with the parameters `r` ( number of
 failures until the experiment is stopped, $r \in N$) and `p` (success
 probability in each experiment, $p \in [0, 1]$), i.e.  

 $$\text{cf(t)} = \text{cfN_NegativeBinomial(t,r,p)} = p^r(1 - (1-p)e^{it}))^{-r};$$ 

 For more details see WIMMER and ALTMANN (1999), also WIKIPEDIA:
 <https://en.wikipedia.org/wiki/Negative_binomial_distribution>

$\textbf{a) EXAMPLE: CF of the NegativeBinomial distribution, r = 5, p = 0.3}$

```r
r <- 5
p <- 0.3
t <- seq(-15, 15, length.out = 1001)
plotReIm(function(t)
        cfN_NegativeBinomial(t, r, p), t,
        title = "CF of the Negative Binomial distribution with r = 5, p = 0.3")
```

$\textbf{b) EXAMPLE: PDF/CDF of the compound NegativeBinomial-Exponential distribution}$

```r
r <- 5
p <- 0.3
lambda <- 5
cfX <- function(t)
        cfX_Exponential(t, lambda)
cf <- function(t)
        cfN_NegativeBinomial(t, r, p, cfX)
x <- seq(0, 10, length.out = 101)
prob <- c(0.9, 0.95, 0.99)
options <- list()
options$isCompound <- TRUE
result <- cf2DistGP(cf, x, prob, options)
```

 <h2>POISSON DISTRIBUTION</h2>

`cfN_Poisson}(t,lambda)` evaluates the characteristic function $\text{cf(t)}$ of the
 Poisson distribution with the rate parameter $\lambda > 0$, i.e. 

 $$\text{cf(t)} = \text{cfN_Poisson(t,lambda)} = e^{\lambda(e^{it}-1)}$$ 

 For more details see WIMMER and ALTMANN (1999), also WIKIPEDIA:
 <https://en.wikipedia.org/wiki/Poisson_distribution>.

 $\text{cfN_Poisson(t,lambda,cfX)}$ evaluates the compound characteristic
 function 

 $$\text{cf(t)} = \text{cfN_Poisson(-1i*log (cfX(t)),lambda)},$$
 where $\text{cfX}$ is function handle of the characteristic function $\text{cfX(t)}$ of a
 continuous distribution and/or random variable $X$. 

 $\textbf{a) EXAMPLE: CF of the Poisson distribution with lambda = 10}$

```r
lambda <- 10
t <- seq(-10, 10, length.out = 501)
plotReIm(function(t)
        cfN_Poisson(t, lambda), t,
        title = "CF of the Poisson distribution with the parameter lambda = 10")
```

$\textbf{b) EXAMPLE: CF compound Poisson & Exponential distribution}$

```r
lambda1 <- 10
lambda2 <- 5
cfX <- function(t)
        cfX_Exponential(t, lambda2)
t <- seq(-10, 10, length.out = 501)
plotReIm(function(t)
        cfN_Poisson(t, lambda1, cfX), t,
        title = "CF of the compound Poisson-Exponential distribution")
```

$\textbf{c) EXAMPLE: PDF/CDF of the compound Poisson-Exponential distribution}$

```r
lambda1 <- 10
lambda2 <- 5
cfX <- function(t)
        cfX_Exponential(t, lambda2)
cf <- function(t)
        cfN_Poisson(t, lambda1, cfX)
x <- seq(0, 8, length.out = 101)
prob <- c(0.9, 0.95, 0.99)
options <- list()
options$isCompound <- TRUE
result <- cf2DistGP(cf, x, prob, options)
```

<h2>POLYA-EGGENBERGER DISTRIBUTION</h2>

`cfN_PolyaEggenberger}(t,a,b,m)` evaluates the characteristic function
 $\text{cf(t)}$ of the Polya-Eggenberger distribution, with the parameters `a` ($a\in \mathbb{R}$), `b` ($b \in \mathbb{R}$), and `m` ($m\in \mathbb{Z}$ ), i.e.  

 $$\text{cf(t)} = \text{cfN_PolyaEggenberger(t,a,b,m)} = {}_{2}F_1\left(-m,a,a+b,1-e^{it}\right);$$

 where ${}_{2}F_1$ denotes the Gauss hypergeometric function. For more details see
 WIMMER and ALTMANN (1999), p. 525. 

$\textbf{a) EXAMPLE: CF of PolyaEggenberger distribution, a=2.2, b=3.3, m=4}$

```r
a <- 2.2
b <- 3.3
m <- 4
t <- seq(-15, 15, length.out = 1001)
plotReIm(function(t)
        cfN_PolyaEggenberger(t, a, b, m), t,
        title = "CF of the Polya-Eggenberger distribution with a = 2.2, b = 3.3, m = 4")
```

$\textbf{b) EXAMPLE: CF of the compound PolyaEggenberger-Exponential distribution}$

```r
a <- 2.2
b <- 3.3
m <- 4
lambda <- 5
cfX <- function(t)
        cfX_Exponential(t, lambda)
t <- seq(-50, 50, length.out = 501)
plotReIm(function(t)
        cfN_PolyaEggenberger(t, a, b, m, cfX), t,
        title = "CF of the compound Polya-Eggenberger-Exponential distribution")
```

$\textbf{c) EXAMPLE: PDF/CDF of the compound PolyaEggenberger-Exponential distribution}$

```r
a <- 2.2
b <- 3.3
m <- 4
lambda <- 5
cfX <- function(t)
        cfX_Exponential(t, lambda)
cf <- function(t)
        cfN_PolyaEggenberger(t, a, b, m, cfX)
x <- seq(0, 2.5, length.out = 101)
prob <- c(0.9, 0.95, 0.99)
options <- list()
options$isCompound <- TRUE
result <- cf2DistGP(cf, x, prob, options)
```
